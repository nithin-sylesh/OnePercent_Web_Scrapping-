```python
for c_link in websites_of_courses:
    request = requests.get(c_link)
    soup = BeautifulSoup(request.content, 'html.parser')
    break
```

```python
dict_chap = {}
for i in soup.find_all('div', style='width:50%;float:left'):
    for j in i.find_all('p'):
        for k in j.find_all('a'):
            name = k.get('name')
            link = k.get('href')
#             print(name, link)
            if name is not None:
                dict_chap[name] = []
                prev = name
            else:
                name = prev
                if link is not None:
                    dict_chap[name].append(link)
for i in soup.find_all('div', style='width:50%;float:right'):
    for j in i.find_all('p'):
        for k in j.find_all('a'):
            name = k.get('name')
            link = k.get('href')
#             print(name, link)
            if name is not None:
                dict_chap[name] = []
                prev = name
            else:
                name = prev
                if link is not None:
                    dict_chap[name].append(link)
```
